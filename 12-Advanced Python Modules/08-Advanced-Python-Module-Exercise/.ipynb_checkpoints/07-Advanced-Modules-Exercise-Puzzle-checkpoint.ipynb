```python
import re
```

```python
def scan(path):
    phone = re.compile(r'\d{3}-\d{3}-\d{4}')
    for root, dirs, files in os.walk(path):
        for f in files:
            with open(root+'/'+f,'r') as content:
                match = re.findall(phone,content.read())
                for m in match:
                    print(f"match in {path}/{root}/{f}: {m}")
        for d in dirs:
            scan(f'{path}/{root}/{d}')
```
